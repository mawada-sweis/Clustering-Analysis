```python
#libraries used
import pandas as pd
```

```python
#load data 
data = pd.read_csv("../data/raw/Citywide_Survey.csv")
```

```python
#disply data
data
```

### Grouping by Houshold Id

```python
household_grouped = data.groupby("hh_id").agg(set)
```

```python
# display result
household_grouped
```

### There are many features (columns) that have values just for one member of the household and the other members have values ( 0 or nan or 995 etc) which actually must be the same value as the member with the  real value. 

```python
# grouped data without nulls
without_null = data.groupby('hh_id').agg(lambda cell: set(cell.dropna()))
```

```python
#display result
without_null
```

### The result of grouping by hh_id ==> the number of the households equal 3346

### Groubing by person_id
