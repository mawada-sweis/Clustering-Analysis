# Split the features based on the type of data
which are: objects, dates, encoded, numeric and labeled features

```python
date_features = ['first_travel_date','last_travel_date']
categorical_features = ['cms_zone', 'work_cms_zone']
encoded_features, numerical_features, labeled_features = CLD.get_features_based_type(data, data.columns)
```

We detected "jobs count" feature in labeled features list instead of "numerical features" by evaluating the features of numeric and labeled lists, thus we will handle this case:

```python
error_numeric_features = ['jobs_count']

# Remove the error features from labeled_features and add them to numeric_features
labeled_features = list(set(labeled_features).symmetric_difference(set(error_numeric_features)))
numerical_features.extend(error_numeric_features)

CLD.display_info_lists(data, ("numerical", numerical_features), ("labeled", labeled_features))
```

# Feature Engineering

```python
# ordinal features (label-encoded)
ordinal_features = ['tnc_freq', 'telework_freq', 'planning_apps', 'drive_cbd_freq', 'congestion', 'english_proficiency', 'bike_freq', 'education', 'age']

# Remove the ordinal features from labeled_features 
labeled_features = list(set(labeled_features).symmetric_difference(set(ordinal_features)))
```

```python
# Encoding labels features by on-hot encoding method
data = TRD.one_hot_encoding(data, labeled_features)
```

```python
# Show data after applying one hot encoding
data.head()
```

```python
# Encoding categorical features by on-hot encoding method
data = TRD.one_hot_encoding(data, categorical_features)
```

```python
# Show data after applying one hot encoding
data.head()
```

# Scaling

```python
# Load transformed data
data_to_scale_min_max= pd.read_csv('../data/processed/data_transformed.csv')
data_to_scale_standard= pd.read_csv('../data/processed/data_transformed.csv')
```

```python
# date features to scale
date_features = ['last_travel_date_month', 'last_travel_date_day', 'first_travel_date_month', 'first_travel_date_day']
```

```python
# list all features to scale 
features_to_scale = numerical_features + ordinal_features + date_features
```

```python
# scale the data with standard scaler 
TRD.scaling(data_to_scale_standard, features_to_scale, StandardScaler())
```

```python
# scale the data with min-max scaler 
TRD.scaling(data_to_scale_min_max, features_to_scale, MinMaxScaler())
```
